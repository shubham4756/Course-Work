```python
import pandas as pd

jdDataSet = pd.read_excel('jd_dataset_skills.xlsx')
jdDataSet.head()
```

```python
import operator

# to do classification only skilss
def extractSkills(string):
    if isinstance(string,float):
        return [""]
    comma_lst = string.split(',')
    comma_lst = [s.strip(" ") for s in comma_lst]
    
    backslash_lst = string.split('/')
    backslash_lst = [s.strip(" ") for s in backslash_lst]
    
    if len(comma_lst) >= len(backslash_lst):
        return comma_lst
    else:
        return backslash_lst

# to generate N grams
def generate_N_grams(text,ngram=1):
    temp=zip(*[text[i:] for i in range(0,ngram)])
    ans=[' '.join(ngram) for ngram in temp]
    return ans

mp = {}

for skills in jdDataSet['skills']:
    skills_list = extractSkills(skills)
    skills_list = [ skill.lower() for skill in skills_list]
    skills_list = skills_list + generate_N_grams(skills_list,2) + generate_N_grams(skills_list,3)
    for skill in skills_list:
        if (skill in mp):
            mp[skill] += 1
        else:
            mp[skill] = 1 
            
# mp = dict(sorted(mp.items(), key=lambda item: item[1]))

# Sort Dictionary by value using itemgetter
sorted_dict = dict(sorted(mp.items(), key=operator.itemgetter(1), reverse=True))

sorted_dict
```

```python
import pandas as pd
df1 = pd.DataFrame(data=sorted_dict,index=[0])
df1 = (df1.T)
df1.to_excel("skills.xlsx")
```
